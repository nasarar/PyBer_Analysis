```python
#magic function
%matplotlib inline
```

```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

## Creating and annotating  line charts

###### line graph  MatLab (PyPlot) interface 

```python
plt.plot(x_axis, y_axis)

#create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label = 'Boston')

#create labels for the x and y axes
plt.xlabel('Date')
plt.ylabel('Fare ($)')

#set the y limit from 0 to 45
plt.ylim(0, 45)

#create title
plt.title('PyBer Fare by Month')

#add grid
plt.grid()

#add legend
plt.legend()

```

###### line graph using Object Oriented interface

```python
#create the plots with ax.plt(); object oriented interface
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

##  Creating Bar Graphs

###### bargraphs using pyplot interface

```python
#create the plot
plt.bar(x_axis, y_axis)
```

```python
# Annonating the Bar chart

#create the plot
plt.bar(x_axis, y_axis, color = 'green', label = 'Boston')

#create the labels for x and y axes
plt.xlabel('Date')
plt.ylabel('Fare ($)')

#create title
plt.title('PyBer Fare by Month')

#add legend
plt.legend()
```

###### horizontal bar graphs using matplot interface

```python
#create horizontal plot
plt.barh(x_axis, y_axis, color = 'magenta', label = 'Boston')

plt.xlabel('Date')
plt.ylabel('Fare ($)')

#add the legend() function in order to display the label created ie. 'Boston'
plt.legend()
#this line inverts the months using the gca() method starting in Dec at the top to starting in Jan
plt.gca().invert_yaxis()
```

###### creating barcharts using the object-oriented interface

```python
#create the plot with ax.plt()
fig, ax = plt.subplots()

#annotates the graph to display colour to black and add label 'New York'
ax.bar(x_axis, y_axis, color = 'black', label = 'New York')
ax.legend()
```

###### creating horizontal barcharts using object-oriented interface

```python
#create the plot with ax.plt()
fix, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = 'cyan', label = 'Chicago')

#annotates the graph
ax.invert_yaxis()
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')

#legend() = label
ax.legend()
```

## Creating Scatter Plots and Bubble Charts

###### Scatter plot using MatLab interface

```python
#two ways to create a scatter plot: plt.plot() vs plt.scatter()

#plt.plot(): where 'o' is the marker and can be changed; without this parameter, this would display a line chart
#y and x axes are switches places due to type of data provided and how the data wants to be visualized
plt.plot(y_axis, x_axis, 'o', color = 'red', label = 'Chicago')

#annotates the scatter plot
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.legend()

#inverts the y-axis so january is on top
plt.gca().invert_yaxis()
```

```python
#plt.scatter()
plt.scatter(y_axis, x_axis, color = 'green', label = 'Chicago')

#annotates the scatter plot
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.legend()

#inverts the y-axis so january is on top
plt.gca().invert_yaxis()
```

###### bubble charts using MatLab interface

```python
#create the chart
#Where the 's' parameter is considered as the size of the markers/bubble
plt.scatter(x_axis, y_axis, s = y_axis)

```

```python
#The size of the markers are still too small so a loop will be used to multiply the size *3
y_axis_larger = []

for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s = y_axis_larger)
```

```python
#List Comprehension can be used instead of using 'for' or 'while' loops to create a more efficient process

plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

###### scatter plot using the object-oriented interface

```python
#create the scatter plot
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis,)
```

###### bubble chart using the object-oriented interface

```python
#create the bubble chart
fig, ax = plt.subplots()

ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis] , c = 'skyblue', alpha = 0.8, edgecolor = 'black', linewidth = 2, label = 'Boston')

#anontate the chart
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
ax.set_title('PyBer Fare by Month')
ax.legend()

#inverts y axis
ax.invert_yaxis()

#limits x axis data
ax.set_xlim(0,50)
```

## Creating Pie Charts

###### pie charts using MatLab interface

```python
#create the pie chart
plt.subplots(figsize=(8,8))
plt.pie(y_axis, labels = x_axis)
```

```python
#explode values on pie chart
plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode = explode_values, labels = x_axis, autopct = '%.1f%%')
```

```python
#Assign 12 colours, one for each month

colors = ['slateblue', 'magenta', 'lightblue', 'green', 'yellowgreen',  "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.pie(y_axis, explode = explode_values, colors = colors, labels = x_axis, autopct = '%.1f%%')

plt.show()
```

###### pie charts using object-oriented interface

```python
#create a pie chart
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ['slateblue', 'magenta', 'lightblue', 'green', 'yellowgreen',  "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
fig, ax = plt.subplots(figsize=(8,8))


ax.pie(y_axis, explode = explode_values, labels = x_axis, shadow = True, autopct = '%.1f%%', colors = colors, startangle = 265)

ax.invert_yaxis()
```

## Adding Error Bars and Changing Ticks

###### Error Bars to a line chart using Matlab interface

```python
import statistics
```

```python
#Get the standard deviation of the values in the y_axis
stdev = statistics.stdev(y_axis)
stdev
```

```python
#Added the x and y axes, and yerr = stdev to the errorbar function
#where yerr calls for showing the error upon the y axis
plt.errorbar(x_axis, y_axis, yerr =  stdev)
```

```python
plt.errorbar(x_axis, y_axis, yerr =  stdev, capsize = 3)
```

###### Error bars to a line chart using the object-oriented interface

```python
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr = stdev, capsize = 3)
plt.show
```

###### Error Bars to a bar chart using Matlab interface

```python
plt.bar(x_axis, y_axis, yerr = stdev, capsize = 3)
```

###### Error Bars to a bar chart using object-oriented interface

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr = stdev, capsize = 3)
plt.show()
```

###### change the major ticks using MatLab interface

```python
#xticks() function to adjust the ticks on the x_axis

#import numpy since its a numerical statistics library that can easily create array or matrices of numbers
import numpy as np

plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step = 5.0))
plt.gca().invert_yaxis()
```

###### change the major ticks using the object-oriented approach

```python
#create the plot and change the ticks to 5.0

fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step = 5.0))
ax.invert_yaxis()
plt.show
```

###### add minor ticks using MatLab interface

```python
from matplotlib.ticker import MultipleLocator

#increase the size of the plot figure
fig, ax = plt.subplots(figsize = (8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step = 5.0))
ax.invert_yaxis()

#create the minor ticks at the increment of 1
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import os

# File to Load (Remember to change these)
city_data_to_load = os.path.join("Resources", "city_data.csv")
ride_data_to_load = os.path.join("Resources", "ride_data.csv")

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#Gets the total rides for each city type
total_rides_per_type = pyber_data_df.groupby(['type']).count()['ride_id']

#total_rides_per_type
```

```python
#Gets the total drivers for each city type
total_drivers_per_type = city_data_df.groupby(['type']).sum()['driver_count']

#total_drivers_per_type
```

```python
#Gets the total amount of fares for each city type
total_fares_per_type = pyber_data_df.groupby(['type']).sum()['fare']

#total_fares_per_type
```

```python
#Gets the average fare per ride for each city type. 
average_fares_per_ride = total_fares_per_type / total_rides_per_type

#average_fares_per_ride
```

```python
#Gets the average fare per driver for each city type. 
average_fares_per_driver = total_fares_per_type / total_drivers_per_type 

#average_fares_per_driver
```

```python
#Creates a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({
      'Total Rides': total_rides_per_type,
      'Total Drivers': total_drivers_per_type,
      'Total Fares': total_fares_per_type,
      'Average Fare per Ride': average_fares_per_ride,
      'Average Fare per Driver': average_fares_per_driver})
#pyber_summary_df
```

```python
#Cleaning up the DataFrame. Deletes the index name
pyber_summary_df.index.name = None
```

```python
#Formats the columns.
pyber_summary_df['Total Rides'] = pyber_summary_df['Total Rides'].map('{:,}'.format)
pyber_summary_df['Total Drivers'] = pyber_summary_df['Total Drivers'].map('{:,}'.format)
pyber_summary_df['Total Fares'] = pyber_summary_df['Total Fares'].map('${:,.2f}'.format)
pyber_summary_df['Average Fare per Ride'] = pyber_summary_df['Average Fare per Ride'].map('${:,.2f}'.format)
pyber_summary_df['Average Fare per Driver'] = pyber_summary_df['Average Fare per Driver'].map('${:,.2f}'.format)

#display dataframe
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
#Reads the merged DataFrame
pyber_data_df
```

```python
#Creates a new DataFrame showing the sum of the fares for each date where the indices are the city type and date.


type_date_df = pyber_data_df.groupby(['date', 'type']).sum()[['fare']]
type_date_df
```

```python
#Resets the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
type_date_df = type_date_df.reset_index()

```

```python
#Creates a pivot table with the 'date' as the index, the columns ='type', and values='fare' to get the total fares for each type of city by the date. 


type_date_pivot = type_date_df.pivot(index = 'date', columns = 'type', values = 'fare')
type_date_pivot
```

```python
#Creates a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.

jan_april_df = type_date_pivot.loc['2019-01-01' : '2019-04-29']
```

```python
#Sets the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.


jan_april_df.index = pd.to_datetime(jan_april_df.index)
```

```python
#Checks that the datatype for the index is datetime using df.info()
jan_april_df.info()
```

```python
#Creates a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
jan_april_df = jan_april_df.resample('W').sum()

jan_april_df
```

```python
#Plots the resample DataFrame using the df.plot() function. 
# Import the style from Matplotlib.
from matplotlib import style

jan_april_df.plot(figsize = (12,5))

#add the titles
plt.title('Total Fare by City Type')
plt.ylabel('Fare ($USD)')
plt.xlabel(None)

#Uses the graph style fivethirtyeight.
style.use('fivethirtyeight')

#saves figure in Analysis Folder
plt.savefig('Analysis/Pyber_fare_summary.png')
```
